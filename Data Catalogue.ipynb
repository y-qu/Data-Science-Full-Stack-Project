```python
from pathlib import Path

def list_files_in_directory(directory):
    data_catalog = []
    directory_path = Path(directory)
    for file_path in directory_path.rglob('*'):
        if file_path.is_file():
            if file_path.exists(): # Replace 'path' with 'file_path'
                try:
                    file_info = {
                        'filename': file_path.name,
                        'path': str(file_path),
                        'size': file_path.stat().st_size
                    }
                    data_catalog.append(file_info)
                except Exception as e:
                    print(f"Could not access file: {file_path}. Error type: {type(e).__name__}, Error message: {str(e)}")
    return data_catalog

directories_to_scan = ['X:/']
all_files = []
for directory in directories_to_scan:
    all_files.extend(list_files_in_directory(directory))

for file_info in all_files:
    print(file_info)
```

```python
import pandas as pd

df = pd.DataFrame(all_files)
df.head(50) # Display the first 5 rows
```

```python
df.to_csv('file_catalog.csv', index=False)
```
